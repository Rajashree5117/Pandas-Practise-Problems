```python
import pandas as pd
df=pd.read_csv('name.csv')
df
#if we want to skip any particular row we can do that by using 'skiprows=row number which means u will skip all rows froom 1 to row number exclusively which means till row number -1 th row '
df=pd.read_csv('name.csv',skiprows=8)
df
#indexing starts from 0 hence header=1 really means row#2 in csv file
df=pd.read_csv('name.csv',header=1)
df
#now for eg we don't have header at all like no sno, name, birthdate then what happens is the first row is considerd as header so as to avoid it we can do as, and automatically it generates 0,1,2 as column names
df=pd.read_csv('name.csv',header=None)
df
#we can provide column names as df=pd.read_csv('xyz.csv',header=None,names=["t","u","v"]) so here column names is t,u,v
#if we want to read only few rows(excluding header) from csv file into dataframe then we can do as
df=pd.read_csv('name.csv',nrows=2)
df
#whenever our csv file has notavalible or na values or some unnecessary values which when encountered u will need to convert it as nan then this is how it is  na_values=[list of values to replace with nan]
df=pd.read_csv('name.csv',na_values=["notavalible","na"])
df
#now let us consider a situation where in a csv file whenever we encounter some unesearry values we will convert to nan, but for some columns the value may be unnecessary , but for another column the same unnecessary value may be useful and its obvio that u don't want to convert that to nan so what we can do is mention the columnname and replacing value
df=pd.read_csv("name.csv",na_values={'SNO':['notavalible','na']})
df
#sometimes we may want to convert any file to .csv then we can do as
df.to_csv('ns.csv')
#index=False which means the index will start from 0 default but when we say index=False 
df.to_csv('nn.csv',index=False)
#now we have one .csv file from there we need to only send some columns to another .csv file then mention columns=[column names]
df.to_csv('nn.csv',columns=['name','bd'])
#sometimes we would not like to export the header to another .csv file then 
df.to_csv('nm.csv',header=False)
```

```python
#Reading excel file 
df=pd.read_excel("ne.xlsx",'Sheet1')
df
#in any column if we have na then we want to convert that na into some value which can be done using defining user defined function
def cp(cell):
    if cell=="na":
        return 'rk'
    return cell
df=pd.read_excel("ne.xlsx",'Sheet1',converters={'name':cp})
df
```

```python
#writing into excel file
#startrow,startcol which tells that skip the first row and 2 columns in beggining while writing into excel
df.to_excel('name.xlsx',sheet_name='Sheet1',startrow=1,startcol=2)
#now we have 2 dataframes but we need to write the 2 dataframes into single excel file but within 2 sheets 
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
