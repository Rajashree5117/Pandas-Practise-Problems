```python
#you can absolutely merge two DataFrames of unequal length in pandas.
#In fact, pd.merge() is specifically designed to handle that gracefully.
#💡 Key idea:
#When you merge two DataFrames, pandas aligns rows based on a common key column (like "city"), not on how many rows each DataFrame has.
#So the lengths don’t need to match at all.

🔹 Example 1 — Unequal lengths
import pandas as pd

# First dataframe (3 rows)
r = {"city": ["Darjeeling", "Manali", "Meghalaya"],
     "ttt": [38, 40, 30]}
rdf = pd.DataFrame(r)

# Second dataframe (only 2 rows)
s = {"city": ["Manali", "Darjeeling"],
     "humidity": [10, 11]}
sdf = pd.DataFrame(s)

# Merge on 'city'
df = pd.merge(rdf, sdf, on="city", how="outer")   # 'outer' keeps all rows
print(df)
Output:
        city  ttt  humidity
0  Darjeeling   38      11.0
1       Manali   40      10.0
2    Meghalaya   30       NaN


👉 Even though rdf has 3 rows and sdf has 2, the merge works fine.
NaN appears where there’s no matching value.

🔹 Merge types (important)
how	Keeps rows from	Missing data behavior
"inner"	Only common keys	Keeps only matching rows
"left"	Left DataFrame	Missing data filled with NaN
"right"	Right DataFrame	Missing data filled with NaN
"outer"	Both DataFrames	Keeps all rows from both sides

🔹 Example 2 — “Left” merge
If you want only rows from the first DataFrame:
df = pd.merge(rdf, sdf, on="city", how="left")
print(df)
Output:

        city  ttt  humidity
0  Darjeeling   38      11.0
1       Manali   40      10.0
2    Meghalaya   30       NaN


🔹 Example 3 — No matching column name?
You can also merge on different column names:
pd.merge(df1, df2, left_on="city_name", right_on="place")

✅ Summary

✔ You can merge DataFrames of different lengths.
✔ pandas aligns rows by matching key values, not by row count.
✔ Use how='outer' to keep all data from both DataFrames.
```

```python
#how="outer" means it performs union between 2 dataframes and gives the merged op of 2dataframes
dss=pd.merge(rdf,sdf,on="city",how="outer")
dss
ds=pd.merge(rdf,sdf,on="city",how="inner")
ds
d=pd.merge(rdf,sdf,on="city",how="left")
d
df=pd.merge(rdf,sdf,on="city",how="right")
df
dff=pd.merge(rdf,sdf,on="city",how="right",indicator=True)
dff
```
