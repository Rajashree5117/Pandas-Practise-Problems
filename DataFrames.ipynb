```python
#DataFrames is a main object in pandas,it is used to represent data with rows and columns
import pandas as pd
#now we have some dataset in some location , to convert the dataset into dataframe we do the below operation ".read_csv("and mention the path , remember in py if the path contains \ it treats as escape sequence , so to avoid that we say r"and mention the entire path"
df=pd.read_csv(r"D:\UPSKILL 2025\Python Packages\bwf_historic_rankings.csv",low_memory=False)
df
```

```python
#when we have our data in dictionary format,but wanna convert it to dataframe then:
d={'name':["jaihanuman","rajasri","rakshitha","vasavi","badminton"],'type':["god","hb","hb","hb","sport"]}
dff=pd.DataFrame(d)
dff
#df.shape() returns a tuple which contains the number of rows and columns
dff.shape
#When you write df.shape(),
#Python thinks you’re trying to call a function named shape.
#But actually, df.shape returns a tuple, like (5, 2).
#And when you add (), you’re trying to “call” that tuple — which is not callable.
#Hence the error:
#TypeError: 'tuple' object is not callable
#It literally means:
#“You’re trying to use parentheses on something that’s not a function.”
#🧠 Quick analogy
#You want	Correct	Wrong
#Get the number of rows and columns	df.shape	df.shape()
#Get column names	df.columns	df.columns()
#But if it’s a method (function), like viewing top rows	✅ df.head()	❌ df.head
```

```python
#when we want to retrieve the row whose rank is max then we can do is df[df.column name==df.column name.max()]
df[df["rank"]==df["rank"].max()]
```

```python
#in the above column we are able to print the entire row, column names whose rank is max, but now we just wanna display the country's name whose rank is maximum
df['country'][df["rank"]==df["rank"].max()]
df[['country','name']][df["rank"]==df["rank"].max()]
```
