```python
#now u have one csv file , u made it to dataframe, now u wanna transpose the matrix and add 2 new columns to it 
df=pd.read_csv('name of the csv file')
df.set_index('column name',inplace=True)
#now to transpose the dataframe we can do as below
df=df.T
df
#now if we check the df.index and we run it , then we will get to kow that it is an object , but we need to convert it to period index, so as to convert it 
df.index=pd.PeriodIndex(df.index,freq='Q-JAN')
df.index
#now in df.index we have periodindex , which means it has all quarters, now u want to know the start time of every quarter , we can get to know it using map()
df['start_date']=df.index.map(lambda x:x.start_time)
df
df['end_date']=df.index.map(lambda x:x.end_time)
df
```
