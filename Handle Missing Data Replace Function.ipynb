```python
import pandas as pd
import numpy as np
df=pd.read_csv("wind.csv")
df
```

```python
#replace() takes 2 parameters , one is the parameter to be relased, the parameter replacing the old value this replacement works on entire table columns
nf=df.replace(-9999,np.nan)
nf
nf=df.replace([-9999,-88888,-77777],np.nan)
nf
#sometimes we need to replace nan values of only some columns not every column which has nan then this is how it works
nf=df.replace({'temparature':'Nan','windspeed':'Nan','rain':'n'},0)
nf
#You wrote 'Nan' (a string) — but in your DataFrame, the missing values are actually np.nan (the special NumPy "Not a Number" object), not the text 'Nan'.
#➡️ So pandas doesn’t find any actual 'Nan' strings to replace.
#💡 Why this works
#np.nan represents missing values (not a text).
#'Nan' is just a normal string.
#When your column actually contains true NaN values (from reading Excel/CSV), you must use np.nan to target them
nf = df.replace({'temparature': np.nan, 'windspeed': np.nan, 'rain': 'n'}, 0)
nf
```

```python
nf=df.replace({-99999:np.nan,np.nan:'yy'})
nf
#This applies replacements globally (to the whole DataFrame) and not column-wise.
#Also, np.nan can’t be used as a key directly in a replace() dictionary — pandas can’t detect NaNs as keys because np.nan != np.nan.
```

```python
nf = df.replace({
    'windspeed': {-9999: np.nan},
    'rain': {np.nan: 'yy'}
})
nf
#'windspeed': {-99999: np.nan} → only replaces -99999 in the windspeed column with NaN.
#'rain': {np.nan: 'yy'} → only replaces missing values in the rain column with 'yy'.
```

```python
#Since NaN can’t be matched directly in a dictionary, a more reliable way is to use fillna() for the rain column:
df['windspeed'] = df['windspeed'].replace(-99999, np.nan)
df['rain'] = df['rain'].fillna('yy')
nf = df
nf
```

```python
#✅ Summary of both methods
#Method	Works for	Recommended
#replace({...})	Numeric replacements	✅ Good for -99999 → NaN
#fillna()	Missing value replacements	✅ Best for NaN → 'yy'
#So the clean final version would be:
df['windspeed'] = df['windspeed'].replace(-99999, np.nan)
df['rain'] = df['rain'].fillna('yy')
nf = df
nf
```

```python
#sometimes we need to replace the letters beside numbers with "" so as to do it we can take the help of regex
nf=df.replace('[A-Za-z]','',regex=True)
nf 
#the above one replaces globally columns whichever have alphabets in it , except nan ones 
#but sometimes we need to replace the alphabets only in specific columns not in every column this is how we do it
nf=df.replace({'temparature':'[A-Za-z]','windspeed':'[A-Za-z]'},'',regex=True)
nf 
```
