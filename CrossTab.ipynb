```python
#crosstab is a type of table in a matrix format that displays the frequency distribution of the variable
import pandas as pd
df=pd.read_excel("survey.xlsx")
df
```

```python
#.crosstab(df.columnname to be on xaxis,df.columnname to be on yaxis)
pd.crosstab(df.state,df.handedness)
pd.crosstab(df.state,df.handedness,margins=True)
#now in both x , y axis we have only 1 , but what if we want to see 2(attributes in row or column then )
pd.crosstab([df.state,df.sex],[df.handedness,df.age],margins=True)
#if u wanna know the parameters need to be passed into a method then , place ur cursor in between the method name and click shift+tab then u will get the description of the method u require 
pd.crosstab(df.state,df.handedness,normalize='index')
```

```python
import numpy as np
#to find the avg age of female and male handedes
pd.crosstab(df.sex,df.handedness,values=df.age,aggfunc=np.average)
```
