```python
import pandas as pd
iw=pd.DataFrame({"city":["Kurnool","Anantapur","Hyderabad"],
                 "temparature":[38,40,30],
                 "humidity":[10,11,9]})
iw
                 
```

```python
w=pd.DataFrame({"city":["Darjeeling","Manali","Meghalaya"],
                 "temparature":[18,15,20],
                 "humidity":[0,1,9]})
w
      
```

```python
df=pd.concat([iw,w])
df
```

```python
#if we observe the index of df then we can see 0,1,2 and its repeating, we don't want it to happen in that wayy
df=pd.concat([iw,w],ignore_index=True)
df
```

```python
df=pd.concat([iw,w],keys=["fav visited places","fav to be visited places"])
df
```

```python
#the above , we have divided the combined table as fav visited places, fav to be visited places, to find the subsets of each , we can do as:
df.loc['fav visited places']
df.loc['fav to be visited places']
```

```python
#now here in the above dataframes ttdf,tssf the cities are in same order but what if we don't have them in same order
r={"city":["Darjeeling","Manali","Meghalaya"],
                 "ttt":[38,40,30],
                 }
s={"city":["Manali","Darjeeling"],
                 "humidity":[10,11]}
rdf=pd.DataFrame(r)
sdf=pd.DataFrame(s)
df=pd.concat([rdf,sdf],axis=1)
df
#here df is wrong where in oth index we have Darjeeling and manali where both are 2 differnet cities and on city's data is missing here , so its giving wrong data op , so what we can do is 
rsdf=pd.DataFrame(r,index=[0,1,2])
ssdf=pd.DataFrame(s,index=[1,0])
ddf=pd.concat([rsdf,ssdf],axis=1)
ddf
```
