```python
import pandas as pd
#pivot allow us to transform or reshape the dataframe
df=pd.read_csv("weather.csv")
df
```

```python
ddf=df.pivot(index="DATE",columns="CITY")
ddf
```

```python
#pivot table is used to summarize and aggregate data inside dataframe
df=pd.read_csv("weather2.csv")
df
```

```python
#now create a dataframe which shows avg temparature on a single day in cities
df.pivot_table(index="CITY",columns="DATE")
```

```python
df.pivot_table(index="CITY",columns="DATE",aggfunc="sum")
```

```python
df.pivot_table(index="CITY",columns="DATE",aggfunc="count")
```

```python
df.pivot_table(index="CITY",columns="DATE",aggfunc="diff")
```

```python
df.pivot_table(index="CITY",columns="DATE",aggfunc="mean")
```

```python
df.pivot_table(index="CITY",columns="DATE",margins=True)
```

```python
df=pd.read_csv("weather3.csv")
df
#here in the date column the values are not in datetime datatype, they are in string format ,so to convert them to datetime format we can do as
df['DATE']=pd.to_datetime(df['DATE'])
df
type(df['DATE'])
#type(df['DATE'][0])
```
