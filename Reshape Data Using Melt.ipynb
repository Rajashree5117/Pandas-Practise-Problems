```python
import pandas as pd
#melt is used to transform or reshape data
df=pd.read_csv("cities.csv")
df
```

```python
#.melt(dataframe name,id_vars=["columnname"])  
#id_vars=["columnname"]) is the column that u want to represent on x-axis
dd=pd.melt(df,id_vars=["Day"])
dd
#🔹 pd.melt() — What it does
#The melt() function in pandas is used to “unpivot” a DataFrame — that is, it converts data from a wide format to a long format.
#Think of it like turning columns into rows.
#🔹 Parameters explained
#df → your original DataFrame
#id_vars=["Day"] → columns that you want to keep as is (they will not be unpivoted)
#Other columns (not listed in id_vars) → are “melted” into two new columns:
#variable → the original column name
#value → the data from that column

#You use melt() when:
#You want to reshape data for easier plotting or analysis (e.g., with Seaborn or Matplotlib).
#You need a tidy format — where each column is a variable and each row is an observation
dd[dd["variable"]=="Hyderabad"]
#.#.melt(dataframe name,id_vars=["columnname"],var_name="new name of column",value_name="new name of another column") 
dd=pd.melt(df,id_vars=["Day"],var_name="city",value_name="temparatuer")
dd
```
