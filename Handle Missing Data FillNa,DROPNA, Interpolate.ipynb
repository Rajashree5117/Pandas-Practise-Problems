```python
#so sometimes in every column there might be nan columns , then in every columnn nan need not be replaced as 0 specifically, instead each columns nan will be replaced with some other values so as to specify that what we can do is
newdf=df.fillna({'subject':"PowerBI",'no of days':9})
newdf
#👉 pandas replaces NaN values in numeric columns with 9.
#But — your "no of days" column is likely of float type (because NaN is a floating-point value in pandas).
#So internally, pandas does this:
#Original column dtype: float64
#You replaced a missing value with 9 (an integer)
#pandas converts 9 → 9.0 to keep the column dtype consistent (float64) 
#That’s why you see:
#9.0 instead of 9
#⚙️ Why pandas does this:
#NaN itself is a float value (float('nan')), so any column that contains NaN automatically becomes float64.
#Even if you fill it with an integer, the dtype doesn’t change.
#✅ Solutions (if you want pure integers):
#Option 1: Convert column dtype after filling
#newdf["no of days"] = newdf["no of days"].astype(int)
#Now all values will be integers (e.g. 9 instead of 9.0).
#Option 2: Use Int64 (nullable integer type)
#If your column may still contain missing values later, use pandas’ nullable integer type:
#newdf["no of days"] = newdf["no of days"].astype("Int64")
#This allows both integers and NaN together without forcing float dtype.
#✅ Summary:
#Goal	Code
#Keep float (default)	fillna() only
#Force integer	.astype(int)
#Allow NaN + integers	.astype("Int64")
```

```python
#now if we have some na values in column , sometimes we will fill the nan value with the value present just above it , to make it happen , we do as:
#nf=df.fillna(method="ffill")  older version to fill it , we also have bfill()
nf=df.ffill()
nf
#when we have nan values in column then the right,or left column has nan then if we say axis=column then it copies the value from either right or left column value to nan
nf=df.ffill(axis="columns")
nf
#🧩 Why it happened:
#In pandas, axis names are strictly:
#axis=0 → operate down columns (i.e. fill values vertically — one column at a time)
#axis=1 → operate across rows (i.e. fill values horizontally — one row at a time)
#You can also use the string names:
#"index" instead of 0
#"columns" instead of 1
#❌ But you wrote "column" (singular),
#✅ it should be "columns" (plural).
#✅ Correct usage:
#If you want to fill horizontally (from left to right column values):
#nf = df.ffill(axis="columns")
#or equivalently:
#nf = df.ffill(axis=1)
#If you want to fill vertically (from top to bottom rows) — the usual behavior:
#nf = df.ffill(axis="index")
#or simply:
#nf = df.ffill()
#🧠 Quick Summary
#Goal	Correct Code	Explanation
#Fill downward (from above rows)	df.ffill(axis=0) or df.ffill()	Most common usage
#Fill rightward (from left columns)	df.ffill(axis=1) or df.ffill(axis="columns")	Less common, fills across columns

#when we have two down columns empty but we want to fill only one below column with the above column value then we can set limit , if we don't what happens is we have some value in first column as 32 and in next 2 rows in same column there is empty value , if we don't set limit and give ffill then 32 will be filled in nxt columns, but if we set limit=1 only next column will be filled with the above column value
newdf=df.fillna(method="ffill",limit=1)
newdf
```

```python
#interpolate() : is the method which helps when there are so many nan or 0 values but u wanna handle those missing data by giving it an near value bw the values then we use interpolate() this is mostly useful for estimating the missing values with the nearest values
newdf=df.interpolate()
newdf
```

```python
#so in our .csv file, if we have na , then we can drop na values as
newdf=df.dropna()
newdf
#sometimes we would like to drop only those rows where entire row is empty mostly then we can do as
newdf=df.dropna(how="all")
newdf
#now we are not droping the row if it has atleast on non nan column in row , this is how we can do it 
newdf=df.dropna(thresh=1)
newdf
```
